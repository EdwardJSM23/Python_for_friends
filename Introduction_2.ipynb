## Mutable e inmutable

Python tiene dos tipos de objetos, mutables e inmutables. En principio, un objeto inmutable es aquel que no puede cambiar y permanece igual durante todo el código, por el contrario, un objeto mutable es aquel que puede cambiar, es decir que puede crecer, reducirse o cambiar su valor o valores.

Listas vs tuplas
* Lista: Son mutables. sus elementos pueden ser distintos tipos de dato. Longitud variable.
* Tupla: Son inmutables. sus elementos deben ser del mismo tipo de dato. Longitud fija.

# Un poco más de funciones

```python
# Multiple arguments

def numeros(primer_numero, segundo_numero, tercer_numero):  
    suma = primer_numero +segundo_numero + tercer_numero
    print('suma: ', suma)
    multiplicacion = primer_numero * segundo_numero * tercer_numero
    print('multiplicacion: ',multiplicacion)
    return suma,multiplicacion
```

```python
numeros(8,5,3)
```

```python
# default arguments

def potencia(numero, power=2):  
    potencia = numero ** power
    return potencia
```

```python
potencia(6,3)
```

```python
# flexible arguments
# non-named variable-length argument (tuple)

def args_example(*args):  
    print(type(args))
    print('args:',args)
    print('suma:' , sum(args))
```

```python
args_example('gato','perro')
```

```python
def flexible_1(nombre_1,*nombres):
    print(type(nombres))
    return nombre_1,nombres
```

```python
type(flexible_1)
```

```python
flexible_1('Barbie', [3,4],5,'gatos')
```

## Decoradores

Son una función que come funciones o que sus argumentos son otra función.

```python
import time

def timer(funct):
    def wrapper(a,b):
        tic = time.time()
        result = funct(a,b)
        toc = time.time()
        print('Time:'+str((toc-tic)*1000)+'s')
        return result
    return wrapper
```

```python
type(timer) 
```

# Notas
Scope, mro. problema en conputación del punto flotante

```python
@timer        # pythonica
def suma(a,b):
    return a+b
```

```python
suma(8,2)
```

```python
def suma(a,b):
    return a+b

suma_decorada = timer(suma)
suma_decorada(8,2)
```

# Importando Paquetes

* los paquetes en Python se importan con la keyword "import" seguido por el nombre de la librería y de manera opcional se le pone un nickname para que sea más fácil de llamar.

* Python tiene unos paquetes instalados por defecto, si no los tiene, de manera general se instalan con el comando "pip install nombre_de_la_lib" o "conda install nombre_de_la_lib".

```python
# first we invite numpy over : import numpy
# at the same time we give numpy the nickname np : as np
import numpy as np

# here is the data we want to investigate
x = [1,3,6,2,8,4,1]

# now that numpy (np) is here we can ask them to use their mean function
x_mean = np.mean(x)

# then we ask numpy to use its standard deviation function on the data
x_standard_dev = np.std(x)

# finally we will print the result
print(x_mean)
print(x_standard_dev)
```
